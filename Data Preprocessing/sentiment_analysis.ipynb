```python
df = pd.merge(news, stocks, on = 'Date', how = 'right').drop('Crypto', axis = 1)
df
```

```python
from transformers import pipeline, BertForSequenceClassification, BertTokenizer
from textblob import TextBlob
import pandas as pd
import torch
from tqdm import tqdm

# Enable tqdm for pandas apply
tqdm.pandas()

# Check if GPU is available
device = 0 if torch.cuda.is_available() else -1  # 0 for GPU, -1 for CPU

sentiment_model = pipeline("sentiment-analysis", model="yiyanghkust/finbert-tone", device=device)

# sentiment_model = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment", device=device)

df['Title'] = df['Title'].fillna('')  # Fill NaN titles with empty strings

def analyze_sentiment(text):
    # Sentiment classification using Hugging Face
    hf_result = sentiment_model(text)[0]
    sentiment_class = hf_result['label']
    score = hf_result['score']
    
    # Polarity and subjectivity using TextBlob
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    # Return the sentiment analysis result
    return {
        'class': sentiment_class.lower(),
        'score': score,
        'polarity': polarity,
        'subjectivity': subjectivity
    }
```

```python
# Apply the analysis to the Title column
# tdf = df.sample(n = 10)
df['Sentiment'] = df['Title'].apply(analyze_sentiment)

# Expand the result dictionary into separate columns
df_expanded = pd.concat([df.drop(['Sentiment'], axis=1), df['Sentiment'].apply(pd.Series)], axis=1)

# Save the output
df_expanded.to_csv("BNB_sentiment_analysis.csv", index=False)
# df_expanded
df
```
